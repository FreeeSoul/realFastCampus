```python
value_25 = stat[407][4]
value_75 = stat[407][6]
print(value_25, value_75)
```

이전까지는 평균값에 대해 크냐 / 작냐로 단순히 나뉘었다면 이번 강의에서 할거는 그렇게 간단하지는 않다.

```python
data_clean['Grade'] = 0
```

```python
data_clean
```

```python
def get_grade(input):
  if input > value_25:
    if input > value_75:
      output = 2
    else:
      output = 1
  else:
    output = 0
  return output
```

```python
data_grade = pd.DataFrame(columns=['time', 'day', 'grade'])
data_grade
```

```python
for index, item in data_clean.iterrows():
  data_grade = data_grade.append({'time': item['집계시'], 'day': item['요일'], 'grade': get_grade(item[407])}, ignore_index=True)
```

```python
data_grade
```

```python
data_grade_list = data_grade.values.tolist()
```

```python
x_train = [r[:2] for r in data_grade_list]
y_train = [int(r[2]) for r in data_grade_list]
```

```python
y_train
# 여기서 정답데이터는 속하는 군집의 인덱스임 0~2번째 인덱스를 가진 총 3개의 군집이 존재 중
```

```python
import tensorflow as tf
```

```python
 y_one_hot = tf.keras.utils.to_categorical(y_train)
 # one-hot encoding화 해주는 코드
```

```python
print(x_train[:3], y_train[:3], y_one_hot[:3])
# 지금 정답이 [1, 1, 0]이고, one_hot을 [0. 1. 0.], [0. 1. 0.], [1. 0. 0.] 로
# 표현하고 있다.
```

이제 이러한 one hot구조로 텐서플로우를 만들어 보자
# Multinomial CLassification using tensorflow

```python
learning_rate = 1e-2
learning_epochs = 1000
```

```python
# 옵티마이저는 그대로 Gradient Decent 사용
sgd = tf.keras.optimizers.SGD(learning_rate=learning_rate)

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(3, input_shape=(2, ), activation='softmax'))
#출력이 3개임 (one-hot모양이 3개 수 나오는 형태라), 2개 넘는 여러 개를 분류하므로 시그모이드도 여러
#개 이다. 그러므로 softmax를 사용해서 한 번에 처리하도록 해야한다.
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
model.summary()
```

```python
import numpy as np
```

```python
type(np.array(x_train))
```

```python
history = model.fit(np.array(x_train), np.array(y_one_hot), epochs=learning_epochs)
# 보면 입력은 그대로 집계시, 요일인데 출력이 원핫인코딩이다 ★★★ 즉 여기서 정답데이터는 원핫코딩 형태이다. 근데 특이한 것은 모델 예측을 통해 나온 정답데이터는
# 원핫코딩이 아니라 소프트맥스로 확률화까지 된 단계의 값이 나온다. 여기서 내가 직접 원핫코딩을 해줘서 출력할 때 군집이 뭔지 나타내도록 해주면된다.★★★
# 항상 머신러닝에서 하는 거는 Rule을 찾는거고, 여기서는 시그모이드 Rule을 찾는 것이다. --> 정확히는 소프트맥스 Rule을 찾아주는 것 같기도.. 시그모이드 3개 대신 softmax쓰는거니까
# 여기서 Rule은 Hypothesis 함수 모양을 명확히 찾는 것이고, 여기서는 함수가 3개다.
# 선형회귀에서는 여러 개의 직선들을 놓아보면서 점들 간의 거리의 제곱 합이 최소인 것을
#찾았었다.시그모이드 함수 사용하는 이유 : 선형회귀로 하면 극단치들의 값이 그대로 반영됨
#그래서 극단치 값들을 완만하게 값을 가지게끔 해주는 함수모양이 sigmoid로 해줘서 값들의 범위를 0과 1사이로 만들어준다. 그러므로 극단치 값이 최대 1로 완만해짐
# cost를 구할 때 선형회귀함수는 그냥 1차식과 같은 형태라 제곱해서 Cost구해주면 이차함수가 예쁘게
# 나오지만, Classification에서의 H(x)모양은 e가 들어간 시그모이드 형태라서 얘를 제곱해주면
#꼬불모양이 나옴. 그래서 이 제곱한 것에 log계산들을 해주어서 Cost를 만들어주는 것
```

```python
Time = 14 #@param {type: 'slider', min:0, max:23, step:1}
Day = 3 #@param {type: 'slider', min:0, max:6, step:1}
```

```python
time_condition = data_clean['집계시'] == Time
day_condition = data_clean['요일'] == Day
data_clean[time_condition & day_condition]
```

```python
input = [Time, Day]
result = model.predict(np.array([input]))
lisT = ['빠름', '보통', '느림']
Final_RESULT = np.argmax(result[0])
print('정답은' + lisT[Final_RESULT])
```

세 개의 군집 분포가 어떻게 되어있나 보고, H(x) 세 번 반복 방식을 쓴 후 각각의 군집들의 score를 구해주고 그것을 소프트팩스로 확률화 해줌. 그럼 세 개 중에 제일 우위에 점한 군집을 확률 제일 높은거로 알려줌. 그럼 그것을 on-hot 인코딩해서 출력
