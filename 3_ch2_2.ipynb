#4. 가공 더 한 좋은 데이터를 가지고 Linear Regression by Tensorflow하기

```python
import tensorflow as tf
import numpy as np
```

```python
Selected_Date = '2020-01-21' #@param {type: 'date'}
input_date = int(Selected_Date.replace('-', ''))
input_date
```

```python
data_date = data[data['집계일자'] == input_date]
data_date
```

```python
plt.plot(data_date['거리'], data_date['통행시간'], 'r*')
```

```python
data_time = data_date.groupby(['집계일자', '거리'])['통행시간'].mean()
data_time
```

```python
data_date.sort_values(by=['통행시간'], ascending=False)
```

```python
data_out = data_time.reset_index()
data_out
```

```python
# DataFrame형태로 이루어진 것을 list로 바꾸어주기 -> 그래야 학습에 쓰게 조정을 막 해대지
data_list = data_out.values.tolist()
data_list
```

```python
x_train = [int(r[1]) for r in data_list] # 학습할 입력값 x를 리스트에서 가져오는데, 텐서플로우가 작업하기 편하게 정수형으로 바꿔준다.
y_train = [int(r[2]/10) for r in data_list] # 값에 10을 나눠주는 이유: normalize과정인데, 지금 입력값은 두 자리 정수인데 출력값이 세 자리 정수니까
# 차수가 다르면 왜곡이 생길 수 있으므로 이 스케일을 같게 하기위해 두 자리 정수로 맞춰줌 크기 느낌에서.
```

```python
plt.plot(x_train, y_train, 'r*')
```

```python
learning_rate = 1e-8
learning_epochs = 2000
```

```python
sgd = tf.keras.optimizers.SGD(learning_rate=learning_rate)
mse = tf.keras.losses.mean_squared_error

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim=1))
model.compile(loss=mse, optimizer=sgd)
```

```python
model.summary()
```

```python
# train model 고고
history = model.fit(x_train, y_train, epochs=learning_epochs)
```

```python
plt.plot(history.history['loss'])
plt.show()
```
